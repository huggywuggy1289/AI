```python
import pandas as pd
from google.colab import files
import io

# 파일 업로드
uploaded = files.upload()

# 업로드된 파일의 키 확인
uploaded_files = list(uploaded.keys())

# 파일 이름에 (숫자)가 있는지 확인하면서 읽기
found = False
for file_key in uploaded_files:
    if '자동차판매량' in file_key:
        df = pd.read_csv(io.StringIO(uploaded[file_key].decode('cp949')), usecols=['연간 자동차 판매량', '총인구(명)', '1인당 국민 총소득($)'])

        # 열 이름 변경
        df = df.rename(columns={'연간 자동차 판매량':'연간 자동차 판매량', '총인구(명)':'인구', '1인당 국민 총소득($)':'일인당국민소득'})

        # 데이터프레임 출력
        print(df)
        print(df.dtypes)
        found = True
        break

if not found:
    print("업로드된 '자동차판매량' 파일을 찾을 수 없습니다.")
```

```python
# 국가 열을 제외하고 나머지 열들을 숫자로 변환
for i in range(0, len(df)):
    for j in range(1, len(df.columns)):
        df.iloc[i, j] = pd.to_numeric(str(df.iloc[i, j]).replace(",", ""))

# '연간 자동차 판매량', '인구', '일인당국민소득' 열을 모두 정수로 변환

df['연간 자동차 판매량'] = pd.to_numeric(df['연간 자동차 판매량'])
df['인구'] = pd.to_numeric(df['인구'])
df['일인당국민소득'] = pd.to_numeric(df['일인당국민소득'])

# 데이터프레임 출력
print(df)
print(df.dtypes)
```

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# 데이터프레임 생성
data = {'car_sales': [10, 20, 30],
        'population': [100, 200, 300],
        'income_per_capita': [1000, 2000, 3000]}

df = pd.DataFrame(data)

# 모델 생성
model = smf.ols(formula='car_sales ~ population + income_per_capita', data=df)
result = model.fit()

# 결과 출력
print(result.summary())
```
